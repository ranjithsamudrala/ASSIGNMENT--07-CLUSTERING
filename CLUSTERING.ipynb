# Hierarchical Clustering

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch # for creating dendrogram
#p = np.array(df_norm) # converting into numpy array format
z = linkage(df_norm, method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
h_complete = AgglomerativeClustering(n_clusters=5, linkage='average',affinity = "euclidean").fit(df_norm)

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
crime['clust']=cluster_labels # creating a  new column and assigning it to new column
crime
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(df_norm)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
X
```

```python
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors=8)
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

```python
dbscan = DBSCAN(eps=0.475, min_samples=4)
dbscan.fit(X)
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
crime1= pd.concat([crime,cl],axis=1)
crime1
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(crime1['cluster'],crime['UrbanPop'], c=dbscan.labels_)
```

```python
# Standaization the data
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
Y = SS.fit_transform(airlines)
X = pd.DataFrame(Y)
```

#kmeans clustering

```python
from sklearn.cluster import KMeans
```

```python
df= KMeans()
```

```python
inertia =[]
for i in range(1,11):
    df = KMeans(n_clusters=i, random_state=0)
    df.fit(X)
    inertia.append(df.inertia_)
```

```python

plt.plot(range(1,11),inertia)
plt.title("Elbow Method")
plt.xlabel("no.cluster")
plt.ylabel("inertia")
plt.show()
```

```python
import seaborn as sns
d1 = {"kvalue":range(1,11),"inertiavalues":inertia}
d2 = pd.DataFrame(d1)
sns.barplot(x="kvalue",y="inertiavalues",data= d2,)
```
